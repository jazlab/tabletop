```python
import shelve
import shutil

from tabletop_utils.ros import msg_to_dict
from tabletop_utils.shelf_sqlite3 import Sqlite3Shelf
```

```python
shutil.copy("test.db", "copy.db")
```

```python
db = shelve.open("cache", flag="r")
new_db = Sqlite3Shelf("copy.db", flag="r")
```

```python
print(f"len(db) = {len(db)}, len(new_db) = {len(new_db)}")
```

```python
diff = []
for i, (k, v) in enumerate(db.items()):
    if not isinstance(v, list):
        continue
    new_v = new_db[k]
    for j in range(len(v)):
        if msg_to_dict(new_v[j].trajectory_msg) != msg_to_dict(
            v[j].trajectory_msg
        ):
            diff.append((i, j))
            break
len(diff)
```
