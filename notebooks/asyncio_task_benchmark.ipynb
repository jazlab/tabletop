```python
import asyncio
import time
```

```python
async def coros(depth):
    if depth == 0:
        return
    await coros(depth - 1)


async def tasks(depth):
    if depth == 0:
        return
    await asyncio.create_task(tasks(depth - 1))


async def timeit(coro, depth, num_iterations):
    start = time.time()
    for _ in range(num_iterations):
        await asyncio.create_task(coro(depth))
    return time.time() - start
```

```python
coros_avg_time = await timeit(coros, 3, 1000)
tasks_avg_time = await timeit(tasks, 3, 1000)

print(f"coros_avg_time: {coros_avg_time}")
print(f"tasks_avg_time: {tasks_avg_time}")
```
