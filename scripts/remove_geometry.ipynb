```python
import os

import trimesh
from tabletop_utils.mesh import (
    load_geometry,
)
```

```python
mesh_dir = "/root/ws/src/tabletop/ros/tabletop_moveit_config/meshes/static"
mesh_filename = os.path.join(mesh_dir, "rig_v2.dae")
```

```python
mesh = load_geometry(mesh_filename, scale=0.01)
assert isinstance(mesh, trimesh.Scene)
mesh.show()
```

```python
print(f"Number of components: {len(mesh.geometry)}")
key, gripper_mesh = list(mesh.geometry.items())[15]
print(key)
gripper_mesh.show()
```

```python
mesh.show()
```

```python
gripper_mesh.export(os.path.join(mesh_dir, "gripper.dae"))
```

```python
mesh.delete_geometry(key)
dae_bytes = trimesh.exchange.dae.export_collada(list(mesh.geometry.values()))
with open(os.path.join(mesh_dir, "rig_mod.dae"), "wb") as f:
    f.write(dae_bytes)
```
